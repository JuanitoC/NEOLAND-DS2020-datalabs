# Eliminamos las features no necesarias

```python
# Realizamos la eliminación de las features
col_drop = ['Parch','SibSp','Age','Name','Cabin']
titanic_list.drop(col_drop,axis =1 , inplace= True)
```

# Transformación de las features en binaria

```python
# We convert Categorical to boolean for first dataset
categorical_feature = titanic_list.dtypes == "object"
# filter categorical columns
categorical_cols = titanic_list.columns[categorical_feature].tolist()
categorical_cols
```

```python
titanic_list = pd.get_dummies(data=titanic_list,columns=categorical_cols)
```
