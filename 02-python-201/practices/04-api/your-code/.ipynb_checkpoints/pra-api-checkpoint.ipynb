```python
import json
def json_print(json_data, limit=None):
    if isinstance(json_data, (str)):
        json_data = json.loads(json_data)
    nice = json.dumps(json_data, sort_keys=False, indent=3, separators=(',',':'))
    print("\n".join(nice.split("\n")[0:limit]))
    if limit is not None:
        print("[....]")
```

## Ejercicio 2
Queremos saber el número de crímenes violentos que se han producido en Reino Unido en una localización (latitud, longitud) y fecha concretas mediante la seguiente [url](https://data.police.uk/docs/method/crimes-at-location/). Implementad un conjunto de funciones para obtener el número de crimenes producidos en una determinada fecha en una determinada localización. 

- La primera función devolverá la latitud y longitud de una determinada dirección postal mediante la API de geolocalización de [google maps](https://pypi.org/project/googlemaps/1.0.2/). 

- La segunda función devolverá el número de crimenes producidos en una determinada fecha y en una determinada localización (latitud, longitud).  

Usa ambas funciones para obtener el número de crimenes violentos en la dirección **Adelaide St, WC2N 4HZ, London, United Kingdom** en abril del 2018. 

**Nota**: Deberéis registraros a [Google Cloud Platform](https://developers.google.com/maps/documentation/javascript/get-api-key) para obtener las credenciales de la API de googlemaps.

**Respuesta**


```python
from geopy.geocoders import Nominatim
import requests
import pandas as pd

def buscacrimenes(Place,Date):
    
    geolocator = Nominatim(user_agent='Pruebas')
    Address = geolocator.geocode (Place)
    Lat = Address.latitude
    Lon = Address.longitude
    subbuscarcrimenes(Lat,Lon,Date)
    
def subbuscarcrimenes(L,l,d):
    URL = 'https://data.police.uk/api/crimes-at-location?date={}&lat={}&lng={}'.format(d,L,l)
    print(d,L,l)
    Crimes = requests.get(URL)
    # lo convierto en formato json
    Crimes =Crimes.json()
    # Hago un for para crear un dataframe
    CrimeDF = pd.DataFrame(columns=Crimes[1].keys())
    for Crime in Crimes:
        CrimeDF = CrimeDF.append(pd.Series(Crime),ignore_index=True)    
    # filtro por tipo de crimen
    return print(CrimeDF['category'].value_counts())
```

```python
Lugar = 'Adelaide St, WC2N 4HZ, London, United Kingdom'
# Lugar = 'Downing St, London'
# Lugar = 'Stanford Bridge'
#Lugar = 'London Eye'
Fecha = '2018-04'
buscacrimenes(Lugar,Fecha)  
```

Queremos conocer los nombres más frecuentes de los recién nacidos en Barcelona por sexo entre los años 1996 y 2018. Usa la función para obtener lo siguiente:

b) Crear una lista ordenada de mayor a menor, a partir de la lista obtenida en el apartado a), según el número de veces que dicho nombre ha sido utilizado.


Para realizar el ejercicio consultad el portal de datos abiertos del Ayuntamiento de Barcelona mediante la siguiente [url](https://opendata-ajuntament.barcelona.cat/es/). 

**Nota 1**: Consultad como realizar las consultas mediante la API en [la pestaña de Desarrolladores](https://opendata-ajuntament.barcelona.cat/es/desenvolupadors)

**Respuesta**

```python
import requests
# Extraigo los datos como limite pongo 1200, corresponde a 25 resultados x2 (Hombre y mujer) durante 24 alos (1996-2019)
Nombres = requests.get('https://opendata-ajuntament.barcelona.cat/data/api/action/datastore_search?resource_id=e1b5dd1f-a88e-43eb-86d1-d3880d9a6718&limit=1200')
NombresDict1 = Nombres.json()
print(NombresDict1.keys())
# Voy seleccionando partes de la estructura para quedarme solo con lo necesario
NombresDict1=NombresDict1['result']
print(NombresDict1.keys())
NombresDict1=NombresDict1['records']
# Finalmente me quedo con la parte records que es un diccionario, el diccionario lo puedo convertir en DataFrame facilmente
NombresDF = pd.DataFrame(NombresDict1)
# Reviso el tipo de los valores, para este ejercicio necesito que el campo 'Nombre' sea numérico así que procedo a convertirlo
print("Tipos de las variables de NombresDF")
print(NombresDF.dtypes)
NombresDF['Nombre'] = NombresDF['Nombre'].astype(int)
print("Después de haber convertido `Nombre` reviso que sea numérica")
print(NombresDF['Nombre'].dtypes)
```

```python
#Ordeno por el número de veces que se ha puesto el nombre
NombresDF.sort_values(by=['Nombre'],ascending=False,inplace=True)
```

```python
# Quito duplicados quedándome con la primera linea, al haberlo ordenado por numero de manera descendente me quedo 
# con la primera linea que es la que tiene el valor más elevado
NombresDF.drop_duplicates(subset=['Any','Sexe'], keep='first',inplace=True)
```

```python
NombresDF.sort_values(by=['Any','Sexe'],ascending=True,inplace=True)
```

Programad una función que devuelva la fecha y hora de los 10 próximos pases de la estación espacial internacional ([ISS](http://api.open-notify.org)) sobre la Torre Eiffel  (especificada por su **longitud** y **latitud**). La función debe devolver una lista de 10 elementos, cada uno de los cuales debe ser una cadena de caracteres con la fecha y la hora de los pases.

**Respuesta**

```python
Lugar = 'Eifel Tower'
# Lugar = 'Downing St, London'
# Lugar = 'Stanford Bridge'
#Lugar = 'London Eye'
buscaISS(Lugar)  
```

```python

## Utilizo un código muy parecido al ejemplo anterior. Añado una linea para convertir la fecha de TIMESTAMP a DAteTIME

from geopy.geocoders import Nominatim
import requests
import pandas as pd
from datetime import datetime

def buscaISS(Place):
    geolocator = Nominatim(user_agent='Pruebas')
    Address = geolocator.geocode (Place)
    Lat = Address.latitude
    Lon = Address.longitude
    subbuscarISS(Lat,Lon)
    
def subbuscarISS(L,l):
    URL = 'http://api.open-notify.org/iss-pass.json?lat={}&lon={}'.format(L,l)
    print(L,l)
    Passes = requests.get(URL)
    # lo convierto en formato json
    Passes = Passes.json()
    Passes = Passes['response']
    PassesDF = pd.DataFrame(Passes)
    PassesDF['risetime'] = pd.to_datetime(PassesDF['risetime'],unit='s')
    print(PassesDF)
```
