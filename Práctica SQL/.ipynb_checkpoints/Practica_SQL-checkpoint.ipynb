## Tárea 2.- Análisis de datos

```python
import seaborn as sns
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Cargo el dataset
DF = pd.read_excel('DF_SQL.xlsx')
DF.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
# Me quedo con 3 variables principales `annual_percentage_rate`, `booking`, y `number_instalments` y las grafico. 
# `Annual_percentage_rate` puede tener distribución normal.

plt.figure(figsize=(15,5))

sns.distplot(DF['number_instalments'],fit=norm, kde=False,bins=15,ax =plt.subplot(1,3,1) )
sns.distplot(DF['annual_percentage_rate'],fit=norm, kde=False,bins=20,ax =plt.subplot(1,3,2))
plt.title('Gráfico de distribución: Comparativa contra distribución normal',loc='center')
sns.distplot(DF['booking'],fit=norm, kde=False,bins=20,ax = plt.subplot(1,3,3))
plt.show()
```

```python
# Realizo test para comprobar si `annual_percentage_rate` tiene distribución normal.
# a) Shapiro-Wilk Test
from scipy.stats import shapiro
data = DF['annual_percentage_rate']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Shapiro-Wilk Test: Distribución normal')
else:
	print('Shapiro-Wilk Test: Distribución no-normal')
    
# b) D’Agostino’s K^2 Test 
from scipy.stats import normaltest
data = DF['annual_percentage_rate']
stat, p = normaltest(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('D’Agostino’s K^2 Test: Distribución normal')
else:
	print('D’Agostino’s K^2 Test: Distribución no-normal')
```

```python
# Dibujo una matriz de correlación entre las variables se establecen las siguientes correlaciones
# 1.- Correlación positiva (0.42) entre el número de plazos y la cantidad
# 2.- Correlación negativa (-0.14) entre el tipo de interés y el número de plazos.
# 3.- Correlación negativa (-0.25) entre la cantidad reservada y el tipo de interés

DFCorr = DF[['number_instalments','booking','annual_percentage_rate']].corr()
plt.title('Correlación entre variables')
sns.heatmap(DFCorr, annot=True, cmap='Blues')
```

```python
## Procedo a graficar las correlaciones detectadas
plt.figure(figsize=[18,7])

sns.regplot(y='booking', x='number_instalments', data=DF,ax=plt.subplot(1,3,1))
plt.title('1.- (0.42) No de plazos y la cantidad')

sns.regplot(y='annual_percentage_rate', x='number_instalments', data=DF,ax=plt.subplot(1,3,2))
plt.title('2.- (-0.14) el tipo de interés y el No de plazos')

sns.regplot(y='booking', x='annual_percentage_rate', data=DF,ax=plt.subplot(1,3,3),)
plt.title('3.- (-0.25) la cantidad reservada y el tipo de interés')

plt.show()
```

```python
## Intento entender la evolución mensual de los préstamos concedidos (Booking)
# La tendendia es que el volumen de préstamos concedidos aumenta en el período de agosto-octubre.

plt.figure(figsize=[18,7])

sns.lineplot(x='month', y='booking',ci=0,data=DF,color='blue',estimator='sum',ax=plt.subplot(1,3,1))
plt.title('Evolución del importe de préstamos concedidos')


sns.distplot(DF['month'],color='green',ax=plt.subplot(1,3,2),kde=False)
plt.title('Evolución del número de préstamos concedidos')


sns.lineplot(x='month', y='booking',data=DF,ci=0,color='red',estimator='mean',ax=plt.subplot(1,3,3))
plt.title('Evolución de media de importe de préstamos  concedidos')
plt.show()

# Hago un dataframe en el que me aparezcan los datos expuestos en los gráficos
DFMonth_Analysis= pd.DataFrame(DF[['month','booking']].groupby(by='month').agg(['sum','count','mean']))
DFMonth_Analysis['% importe'] = round(DFMonth_Analysis['booking','sum'] / DF['booking'].sum(),2)*100
DFMonth_Analysis['% Ud'] = round(DFMonth_Analysis['booking','count'] / DF['booking'].count(),2)*100


# Durante los últimos 4 meses del año se condede la mayoría del importe y los préstamos.
## llama la atención que en los préstamos solicitados en el período vacacional (meses 6-10) la media del préstamo es más elevada.
print('Durante los últimos 4 meses del año se concede el {} % del importe de todo el año'.format(DFMonth_Analysis.iloc[8:12,3].sum()))
print('Durante los últimos 4 meses del año se concede el {} % de los préstamos de todo el año'.format(DFMonth_Analysis.iloc[8:12,4].sum()))

DFMonth_Analysis.T
```

```python
# Ánalisis de préstamos por tipos.
# Hago un dataframe en el que me aparezcan los datos expuestos en los gráficos
DFIndustryname_Analysis= pd.DataFrame(DF[['industry_name','booking']].groupby(by='industry_name').agg(['sum','count','mean']))
DFIndustryname_Analysis['% importe'] = round(DFIndustryname_Analysis['booking','sum'] / DF['booking'].sum(),2)*100
DFIndustryname_Analysis['% Ud'] = round(DFIndustryname_Analysis['booking','count'] / DF['booking'].count(),2)*100
DFIndustryname_Analysis.sort_values(by='% importe',ascending=False,inplace=True)


print('el Top 3 corresponde al {} % del crédito concedido'.format(DFIndustryname_Analysis.iloc[0:3,3].sum()))
print('el Top 10 corresponde al {} % del crédito concedido'.format(DFIndustryname_Analysis.iloc[0:10,3].sum()))

DFIndustryname_Analysis
```

```python
## Quiero comprobar si el sector del préstamo (Industry_name), afecta a las otras variables como annual_pergentage_Rate, booking o instalments.
## Para ello creo una tabla dinámica
DFCHI = DF.loc[:,DF.columns!='month'].pivot_table(index='industry_name',aggfunc=np.mean)
DFCHI
```

```python
# ejecuto el test chi2, y acepto la hipóteses que el tipo de préstamo afecta a las demás variables.
from scipy.stats import chi2_contingency
table = DFCHI

stat, p, dof, expected = chi2_contingency(table)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
```

```python
# Las variables son independientes así que procedo a hacer un análisis de `Industry_name` segregándolo en 4 categorías.
    # 1.- Top 1 Hogar y Decoración
    # 2.- Top 2 Informatica y Electrónica
    # 3.- Top 3 Educación y Formación
    # 4-  Otros

DF['Industry_name_2'] = 'Otros' 
DF.loc[DF['industry_name']=='Hogar y Decoración','Industry_name_2'] = 'Top 1 Hogar y Decoración'
DF.loc[DF['industry_name']=='Informatica y Electrónica','Industry_name_2'] = 'Top 2 Informatica y Electrónica'
DF.loc[DF['industry_name']=='Educación y Formación','Industry_name_2'] = 'Top 3 Educación y Formación'
DF['Industry_name_2'].unique()
```

```python
# Lanzo las correlaciones de los tipos de préstamo, se mantienen a nivel particular las mismas correlaciones que a nivel general
    # 1.- Correlación positiva (0.42) entre el número de plazos y la cantidad
    # 2.- Correlación negativa (-0.14) entre el tipo de interés y el número de plazos.
    # 3.- Correlación negativa (-0.25) entre la cantidad reservada y el tipo de interés


plt.figure(figsize=[28,5])
for c,v in enumerate(DF['Industry_name_2'].unique()):
    DF.loc[DF['Industry_name_2']==v,['number_instalments','booking','annual_percentage_rate']].corr()   
    
    sns.heatmap(DFCorr, annot=True, cmap='Blues',ax=plt.subplot(1,4,c+1))
    plt.title(['Correlación: ',v])
```

```python
plt.figure(figsize=[18,7])

sns.lineplot(x='month', y='booking',data=DF,ci=0,hue='Industry_name_2',estimator='sum')
plt.title('Evolución del importe de préstamos concedidos')
```
